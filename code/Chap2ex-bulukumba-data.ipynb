```python
import requests
import pandas as pd
import io

url = "https://raw.githubusercontent.com/BugisDev/bulukumba-data/master/pendidikan/statistik_pelajar_berdasarkan_jeniskelamin.csv"
s = requests.get(url).content
csv_data = pd.read_csv(io.StringIO(s.decode('utf-8')))
df = pd.DataFrame(csv_data)
df
```

```python
import thinkstats2
import thinkplot
jenjang = df["Jenjang"]
hist = thinkstats2.Hist(df.Jenjang)
thinkplot.Hist(hist)
thinkplot.Config(xlabel='Jenjang', ylabel='Count', stacked=True, colormap='Paired')
thinkplot.show()
```

```python
laki = df["Laki-laki"]
mean = laki.mean()
var = laki.var()
std = laki.std()
```

```python
mean, std, var
```

```python
jenjang = df[df.Jenjang == 'SD']
hist = thinkstats2.Hist(jenjang.Kecematan)
thinkplot.Hist(hist)
thinkplot.Config(xlabel='Jenjang', ylabel='Count', stacked=True, colormap='Paired')
thinkplot.show()
jenjang.Kecematan.value_counts().sort_index()
```
